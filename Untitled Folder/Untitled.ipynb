# Esempio encoder/decoder
Preso da chatGPT.

Primo esempio di struttura encoder, decoder senza attention mechanism (quindi la versione base con RNN).

## Preprocessing testo 
Cioè trasformo in sequenze ed effettuo padding.

Per l'output, indico le sequenze finali come one-hot encoding.

```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
```

```python
df = pd.read_csv("machine_translation.csv")
df=df.apply(lambda x: "<SOS> " + x + " <EOS>")

df.head()
```

```python


# Esempio di frasi di input e output
# frasi_input = ['Ciao come stai', 'Buongiorno come va']
# frasi_output = ['Hello how are you', 'Good morning how are you']
frasi_input = df["italian"].values
frasi_output = df["english"].values

num_words=2000


# Tokenizer per l'input (italiano)
tokenizer_input = Tokenizer(num_words=num_words)
tokenizer_input.fit_on_texts(frasi_input)
sequences_input = tokenizer_input.texts_to_sequences(frasi_input)
max_len_input = max(len(seq) for seq in sequences_input)
input_data = pad_sequences(sequences_input, maxlen=max_len_input,
                           padding='pre')

# Tokenizer per l'output (inglese)
tokenizer_output = Tokenizer(num_words=num_words)
tokenizer_output.fit_on_texts(frasi_output)
sequences_output = tokenizer_output.texts_to_sequences(frasi_output)
max_len_output = max(len(seq) for seq in sequences_output)
output_data_seq = pad_sequences(sequences_output, maxlen=max_len_output,
                                padding='pre')

print(output_data_seq.shape)

# Output con one-hot encoding
output_data_class = to_categorical(output_data_seq,
                                   num_classes=\
                                   len(tokenizer_output.word_index) + 1)

# Variabili utili
vocab_size_input = len(tokenizer_input.word_index) + 1  # Aggiungiamo 1 per il token zero-padding
vocab_size_output = len(tokenizer_output.word_index) + 1
```

```python
output_data_class.shape
```

## Implementazione encoder/decoder
Senza attenzione

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, Embedding
from tensorflow.keras.backend import clear_session

# Dimensioni degli embedding e unità LSTM (questi sono iperparametri)
embedding_dim = 64
lstm_units = 512

# Encoder
encoder_inputs = Input(shape=(max_len_input,), name="input_encoder")
encoder_embedding = Embedding(input_dim=vocab_size_input, 
                              output_dim=embedding_dim,
                              name="Embedding_input_ita")(encoder_inputs)

encoder_lstm, state_h, state_c = LSTM(lstm_units, 
                                      return_state=True,
                                      name="LSTM_encoder")(encoder_embedding)

# Prendiamo solo gli stati nascosti (state_h, state_c) per passarli al decoder
encoder_states = [state_h, state_c]

# Decoder
decoder_inputs = Input(shape=(max_len_output,), name="input_decoder")
decoder_embedding = Embedding(input_dim=vocab_size_output, 
                              output_dim=embedding_dim,
                              name="Embedding_output_eng")(decoder_inputs)

decoder_lstm = LSTM(lstm_units, 
                    return_sequences=True, 
                    return_state=True,
                    name="LSTM_decoder")

# Passiamo gli stati iniziali dell'encoder come stato iniziale del decoder
decoder_outputs, _, _ = decoder_lstm(decoder_embedding, 
                                     initial_state=encoder_states)

# Strato denso per generare il vocabolario finale
decoder_dense = Dense(vocab_size_output, 
                      activation='softmax', 
                      name="Softmax_decoder")
decoder_outputs_final = decoder_dense(decoder_outputs)

clear_session()

# Creazione del modello Encoder-Decoder completo
model = Model([encoder_inputs, decoder_inputs], decoder_outputs_final)

# Compilazione del modello
model.compile(optimizer='adam', loss='categorical_crossentropy',
              metrics=['accuracy'])

# Mostra il riepilogo del modello
model.summary()
```

Il modello creato su keras è unico, ma in realtà è composto da due parti, cioè encoder e decoder appunto. 

Per questo motivo si definiscono due InputLayer.

## Addestramento modello complessivo

```python
# Addestramento del modello
model.fit([input_data, output_data_seq], output_data_class,
          batch_size=32, epochs=2)
```

```python
# Encoder per inferenza
encoder_model = Model(encoder_inputs, encoder_states)
plot_model(encoder_model, to_file="inference_encoder_model.png", show_shapes=True, 
           show_layer_names=True)
```

```python
# Decoder per inferenza
decoder_state_input_h = Input(shape=(lstm_units,),
                               name="decoder_state_input_h")
decoder_state_input_c = Input(shape=(lstm_units,),
                               name="decoder_state_input_c")
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

# Decoder embedding
decoder_outputs_inference, state_h_inference, state_c_inference =\
decoder_lstm(decoder_embedding, initial_state=decoder_states_inputs)

decoder_hidden_states = [state_h_inference, state_c_inference]

# Decoder finale con lo strato denso
decoder_outputs_final_inference = decoder_dense(decoder_outputs_inference)

decoder_model = Model(
    # input di sequenze + input di stati (dall'encoder)
    [decoder_inputs] + decoder_states_inputs,
    # output della softmax            +   hidden states di decoder
    [decoder_outputs_final_inference] + decoder_hidden_states)
plot_model(decoder_model,to_file="inference_decoder_model.png", 
           show_shapes=True, 
           show_layer_names=True)
```

## Utility function per predizioni nuove


```python
def decode_sequence(input_seq):
    #creazione vocabolario con carattere padding
    dictionary_output=tokenizer_output.index_word.copy()
    dictionary_output[0]="<PAD>"
    
    translated_words=[]
    
    # Codifica l'input come stato iniziale del decoder
    states_value = encoder_model.predict(input_seq)

    #come primo input uso una sequenza di 0.
    #la shape è (1,max_len_output) perchè: ho 1 solo timestamp, di cui
    #faccio padding sulla max_len_output
    target_seq=pad_sequences([[0]], maxlen=max_len_output) #equivalente
#     target_seq = np.zeros((1, max_len_output)) #equivalente

    translated_sentence = ''
    
    #costruisco il testo tradotto una parola per volta (un timestamp
    #per volta). Per farlo devo chiamare ricorsivamente il decoder
    for t in range(max_len_output + 1):
        output_tokens, h, c =\
        decoder_model.predict([target_seq] + states_value)
        
        print(f"output_tokens.shape: {output_tokens.shape}")
        
        #nel costruire la parola step by step, prendo solo la prima
        #della sequenza predetta.
        sampled_token_index = np.argmax(output_tokens[0, 0, :])
        sampled_word = dictionary_output[sampled_token_index]
        translated_words.append(sampled_word)
        
        if sampled_word == 'eos':
            break
        
        #il nuovo input è la sequenza appena predetta (quindi si deve
        #prima passare per argmax), inteso come sequenza di token.
        #poi questi token subiranno di nuovo l'embedding e il resto.
        print(np.argmax(output_tokens,axis=2))
        target_seq = np.argmax(output_tokens,axis=2)
        states_value = [h, c] #questi sono i nuovi stati passati ricorsivamente
    
    return " ".join(translated_words)
```
